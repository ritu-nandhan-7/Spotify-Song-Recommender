```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
dataset['playlist_subgenre'] = label_encoder.fit_transform(dataset['playlist_subgenre'])
dataset['playlist_name'] = label_encoder.fit_transform(dataset['playlist_name'])
dataset['track_artist'] = label_encoder.fit_transform(dataset['track_artist'])


dataset
```

```python
import warnings
warnings.filterwarnings('ignore')


from sklearn_extra.cluster import KMedoids
k_values = range(1, 10)
total_costs = []
```

```python
for k in k_values:
    kmedoids = KMedoids(n_clusters=k, random_state=0)
    kmedoids.fit(dataset)
    total_costs.append(kmedoids.inertia_)  # Sum of distances of samples to their closest cluster center
```

```python
# Plotting the elbow graph
plt.figure(figsize=(8, 4))
plt.plot(k_values, total_costs, 'bo-')
plt.xlabel('Number of clusters, k')
plt.ylabel('Total Cost (Inertia)')
plt.title('Elbow Method For Optimal k in K-Medoids')
plt.grid(True)
plt.show()
```

```python
kmedoids = KMedoids(n_clusters=2, random_state=0)
labels = kmedoids.fit_predict(dataset)
```

```python
data_array = dataset.to_numpy()

medoids = kmedoids.cluster_centers_

plt.figure(figsize=(8, 6))
colors = ['b','y']
labels = kmedoids.labels_

for i in range(len(colors)):
    plt.scatter(data_array[labels == i, 0], data_array[labels == i, 1], s=10, label=f'Cluster {i+1}')
    plt.scatter(medoids[i, 0], medoids[i, 1], s=200, c='k', marker='X', alpha=0.5, label=f'Medoid {i+1}')

plt.title('K-Medoids Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
from scipy.spatial.distance import cdist

def refine_recommendations(base_songs, full_data, top_p):
    """
    Refine recommendations by calculating cosine similarity based on metadata.

    Parameters:
    - base_songs (DataFrame): DataFrame of n-nearest songs selected from the cluster.
    - full_data (DataFrame): Full dataset for additional metadata comparison.
    - top_p (int): Number of final recommendations to return.

    Returns:
    - DataFrame: DataFrame of top p refined recommended songs.
    """
    # Adjust these column names based on your dataset's actual metadata columns
    features = full_data[['playlist_id', 'playlist_genre','playlist_subgenre', 'track_artist']]  # Example column names
    dummies = pd.get_dummies(features)
    base_features = dummies.loc[base_songs.index]
    
    # Calculate cosine similarity
    similarity_matrix = cosine_similarity(base_features, dummies)

    # Get the top p indices for each song and then flatten the list
    top_indices = [similarity_matrix[i].argsort()[-top_p:] for i in range(len(similarity_matrix))]
    unique_indices = list(set([index for sublist in top_indices for index in sublist]))

    # Ensure we do not exceed the number of requested songs and remove potential duplicates
    if len(unique_indices) > top_p:
        unique_indices = unique_indices[:top_p]

    return full_data.iloc[unique_indices]

def recommend_songs(track_id, songs_df, num_nearest_n=10, num_final_p=5):
    """
    Recommend songs based on k-medoids clustering and refined by metadata similarity.

    Parameters:
    - track_id (int): The ID of the song for recommendations.
    - songs_df (DataFrame): DataFrame containing the songs with cluster assignments.
    - num_nearest_n (int): Number of nearest neighbor songs to select initially.
    - num_final_p (int): Number of final recommendations after refinement.

    Returns:
    - DataFrame: DataFrame containing the recommended songs.
    """
    if track_id not in songs_df['track_id'].values:
        return "No song found with track_id {}".format(track_id)

    # Get the cluster ID for the track_id
    cluster_id = songs_df.loc[songs_df['track_id'] == track_id, 'cluster_id'].iloc[0]
    cluster_songs = songs_df[songs_df['cluster_id'] == cluster_id]

    # Calculate distances based on audio features
    audio_features = cluster_songs[['danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness',
                                    'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo',
                                    'duration_ms', 'track_popularity']]  # Update this list as needed
    song_features = audio_features.loc[songs_df['track_id'] == track_id].values.reshape(1, -1)
    distances = cdist(song_features, audio_features, 'euclidean')
    nearest_indices = distances.argsort()[0][:num_nearest_n]
    nearest_songs = cluster_songs.iloc[nearest_indices]

    # Refine these recommendations
    refined_recommendations = refine_recommendations(nearest_songs, songs_df, num_final_p)

    return refined_recommendations

# Directly use the functions with a sample track ID
track_id_input = '5SmYtofSMKeccnqWksS9vW'
recommendations = recommend_songs(track_id_input, df1, num_nearest_n=20, num_final_p=10)
recommendations
```
